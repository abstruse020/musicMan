#Making 2d list

```
def convert(data):
  new_data=[]
  for id_user in users:
    id_songs = data.iloc[:,1][data.iloc[:,0]==id_user]
    #print(id_songs.head())
    id_ratings = data.iloc[:,2][data.iloc[:,0]==id_user]
    #print(id_ratings.head())
    ratings = np.zeros(no_songs)
    if(id_songs.empty==False):
      for sid,rat in zip(id_songs,id_ratings):
        #print(sid)
        indx=songs.index(str(sid))
        ratings[indx] = rat
      new_data.append(list(ratings))
  return new_data
```

#RBM class

```
class RBM():
  def __init__(self,nv,nh):
    #self.W = torch.randn(nh,nv)
    #self.a = torch.randn(1,nh)
    #self.b = torch.randn(1,nv)
    self.W = np.random.randn(nh,nv)
    self.a = np.random.randn(1,nh)
    self.b = np.random.randn(1,nv)
  def sample_h(self,x):
    #wx = torch.mm(x,self.W.t())
    wx = np.matmul(x,self.W.T)
    print('x X W.T =',x.shape,self.W.T.shape)
    #activation = wx + self.a.expand_as(wx)
    activation = wx + self.a
    #p_h_given_v = torch.sigmoid(activation)
    p_h_given_v = self.sigmoid(activation)
    p_h_given_v_bin = (p_h_given_v>0.5).astype(np.int_())
    return p_h_given_v, p_h_given_v_bin #replacing torch.bernoulli(p_h_given_v)
  
  def sigmoid(seld,x):
    return 1./(1. + np.exp(-x))
  
  def sample_v(self,y):
    #wy = torch.mm(y,self.W)
    wy = np.matmul(y,self.W)
    print('y X W =',y.shape,self.W.shape)
    #activation = wy + self.b.expand_as(wy)
    activation = wy + self.b
    #p_v_given_h = torch.sigmoid(activation)
    p_v_given_h = self.sigmoid(activation)
    p_v_given_h_bin = (p_v_given_h > 0.5).astype(np.int_())
    return p_v_given_h, p_v_given_h_bin #torch.bernoulli(p_v_given_h)
  
  def train(self, v0,vk,ph0,phk):
    #self.W += (torch.mm(v0.t(),ph0) - torch.mm(vk.t(),phk)).t()
    #self.b += torch.sum((v0 - vk), 0)
    #self.a += torch.sum((ph0 - phk), 0)
    self.W += (np.matmul(v0.T,ph0) - np.matmul(vk.T,phk)).T
    self.b += np.sum((v0 - vk), 0)
    self.a += np.sum((ph0 - phk), 0)
    
  def returnWeight_W(self):
    return self.W
  def returnWeight_b(self):
    return self.b
  def returnWeight_a(self):
    return self.a
```

```
nv = no_songs
nh  = 100
batch_size = 100
rbm = RBM(nv,nh)
```

#Training the model

```

nb_epoch =20
for epoch in range(1,nb_epoch + 1):
  train_loss =0
  s=0.
  for id_user in range(0,no_users-batch_size,batch_size):
    vk=v0=training_set[id_user:id_user+batch_size]
    ph0,_ = rbm.sample_h(v0)
    for k in range(10): #k step contrastive divergence
      _,hk = rbm.sample_h(vk)
      _,vk = rbm.sample_v(hk)
      vk[v0<0] = v0[v0<0]
    phk,_ = rbm.sample_h(vk)
    rbm.train(v0,vk,ph0,phk)
    #train_loss += torch.mean( torch.abs( v0[v0>=0] - vk[v0>=0] ) )
    train_loss += np.mean( np.abs( v0[v0>=0] - vk[v0>=0] ) )
    s +=1.
  print('epoch: '+str(epoch) + ' loss ' + str(train_loss/s))
```
